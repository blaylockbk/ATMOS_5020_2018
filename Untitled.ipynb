# 1-D plotting with matplotlib

First, need to create some data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.rand(10,10)
x = np.random.rand(10,10)
z = np.random.rand(10,10)
```

```python
y
```

```python
plt.scatter(x, y, c=z)
plt.colorbar()

plt.xlabel('x value')
plt.ylabel('y value')
```

```python
# RGB Image
plt.imshow(np.dstack([y,x,z]))
```
